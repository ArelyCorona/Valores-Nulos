```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('listings.csv')

df.head(5)
```

```python
#ELIMINA COLUMNAS INNECESARIAS
#axis = 1 para columnas
#elimine columnas que no aportan valor analÃ­tico al anÃ¡lisis, 
# como identificadores Ãºnicos o enlaces que no son Ãºtiles para el procesamiento de datos
df2 = df.drop(['id','listing_url','host_is_superhost',], axis =1)
df2.head(5)
```

```python
#IDENTIFICAR VALORES NULOS POR COLUMNA
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#IDENTIFICAR VALORES NULOS POR DATAFRAME
valores_nulos=df2.isnull().sum().sum()
valores_nulos
```

```python
df3=df2.copy()
```

```python
#SUSTITUIR VALORES NULOS CON PROMEDIO O MEDIA
#la columna "bedrooms" la reemplace por el promedio de los valores no nulos
df3["bedrooms"]=df3["bedrooms"].fillna(round(df2["bedrooms"].mean(),1))
```

```python
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#AquÃ­ los valores nulos los reemplace con la mediana. 
df3["reviews_per_month"]=df3["reviews_per_month"].fillna(round(df2["reviews_per_month"].median(),1))

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#TERCER MÃ‰TODO DE SUSTITUCIÃ“N
#SUSTITUIR VALORES NULOS POR UN VALOR NÃšMERICO EN CONCRETO - SE UNA 
#Los valores nulos de la columna "bathrooms" los reemplazacep
# or un valor numÃ©rico fijo (en este caso, 2)

df3["bathrooms"]=df3["bathrooms"].fillna(2)

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#CUARTO MÃ‰TODO DE SUSUTITUCIÃ“N
#SUSTITUIR VALORES NULOS POR UN STRING EN CONCRETO
# los valores nulos los reemplace con un string especÃ­fico ("100%")

df3["host_response_rate"]=df3["host_response_rate"].fillna("100%")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#SUSTITUIR VALORES NULOS POR VALORES NO NULOS HACIA ADELANTE "forward fill"
#FILTRO POR COLUMNAS
#Reemplace los valores nulos con el Ãºltimo valor vÃ¡lido encontrado antes de ellos
df3["beds"]=df3["beds"].fillna(method="ffill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#SUSTITUIR VALORES NULOS POR VALORES NO NULOS HACIA ADELANTE "backward fill"
#FILTRO POR COLUMNAS
#Reemplace los valores nulos con el Ãºltimo valor vÃ¡lido encontrado despues de ellos
df3["host_response_time"]=df3["host_response_time"].fillna(method="bfill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["bathrooms_text"]=df3["bathrooms_text"].fillna(method="bfill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["host_acceptance_rate"]=df3["host_acceptance_rate"].fillna(method="ffill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["price"]=df3["price"].fillna(method="ffill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```
