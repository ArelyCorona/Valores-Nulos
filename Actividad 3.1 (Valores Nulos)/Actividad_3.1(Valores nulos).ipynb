```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ventas_totales.csv')

df.head(5)
```

```python
#ELIMINA COLUMNAS INNECESARIAS
#axis = 1 para columnas
#axis = 0 para filas
df2 = df.drop(['indice_tiempo'], axis =1)
df2.head(5)
```

```python
#IDENTIFICAR VALORES NULOS POR COLUMNA
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#IDENTIFICAR VALORES NULOS POR DATAFRAME
valores_nulos=df2.isnull().sum().sum()
valores_nulos
```

```python
#REALIZAMOS UNA COPIA DEL DATAFRAME
df3=df2.copy()
```

```python
#PRIMER MÃ‰TODO DE SUSUTITUCIÃ“N
#SUSTITUIR VALORES NULOS CON PROMEDIO O MEDIA
#Columna: salon_ventas -MÃ©todo: Se sustituyen los valores nulos con el promedio (media).
df3["salon_ventas"]=df3["salon_ventas"].fillna(round(df2["salon_ventas"].mean(),1))
```

```python
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#SEGUNDO MÃ‰TODO DE SUSUTITUCIÃ“N
#sUSUTITUIR VALORES NULOS CON MEDIANA
#Columna: tarjetas_credito-MÃ©todo: Se sustituyen los valores nulos con la mediana.

df3["tarjetas_credito"]=df3["tarjetas_credito"].fillna(round(df2["tarjetas_credito"].median(),1))

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#TERCER MÃ‰TODO DE SUSTITUCIÃ“N
#SUSTITUIR VALORES NULOS POR UN VALOR NÃšMERICO EN CONCRETO - SE UNA 
#Columna: almacen-MÃ©todo: Se reemplazan valores nulos por 10 y podemos usar cualquuier nÃºmero no solo el 10

df3["almacen"]=df3["almacen"].fillna(10)

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#CUARTO MÃ‰TODO DE SUSUTITUCIÃ“N
#SUSTITUIR VALORES NULOS POR UN STRING EN CONCRETO
#Columna: subtotal_ventas_alimentos_bebidas-MÃ©todo: Valores nulos son reemplazados por el string "21328882.9"
#yo use ese nÃºmero del data 

df3["subtotal_ventas_alimentos_bebidas"]=df3["subtotal_ventas_alimentos_bebidas"].fillna("21328882.9")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#QUINTO MÃ‰TODO DE SUSUTITUCIÃ“N 
#SUSTITUIR VALORES NULOS POR VALORES NO NULOS HACIA ADELANTE "forward fill"
#FILTRO POR COLUMNAS
#Columnas: otros, tarjetas_debito, otros_medios, panaderia, lacteos, carnes, 
# verduleria_fruteria, alimentos_preparados_rotiseria, 
# indumentaria_calzado_textiles_hogar-MÃ©todo: Los valores nulos son sustituidos por el Ãºltimo valor no nulo encontrado.
df3["otros"]=df3["otros"].fillna(method="ffill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#SEXTO MÃ‰TODO DE SUSUTITUCIÃ“N 
#SUSTITUIR VALORES NULOS POR VALORES NO NULOS HACIA ADELANTE "backward fill"
#FILTRO POR COLUMNAS
#Columnas: electronicos_articulos_hogar, bebidas-
# MÃ©todo: Los valores nulos son sustituidos por el siguiente valor no nulo encontrado.
df3["electronicos_articulos_hogar"]=df3["electronicos_articulos_hogar"].fillna(method="bfill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["bebidas"]=df3["bebidas"].fillna(method="bfill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["tarjetas_debito"]=df3["tarjetas_debito"].fillna(method="ffill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["otros_medios"]=df3["otros_medios"].fillna(method="ffill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["panaderia"]=df3["panaderia"].fillna(method="ffill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["carnes"]=df3["carnes"].fillna(method="ffill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["verduleria_fruteria"]=df3["verduleria_fruteria"].fillna(method="ffill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["alimentos_preparados_rotiseria"]=df3["alimentos_preparados_rotiseria"].fillna(method="ffill")

#Comprobamos valores nulos
valores_nulos=df3.isnull().sum()
valores_nulos
```
